```python

from google.colab import drive
drive.mount('/content/drive')


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from matplotlib.ticker import FuncFormatter

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (11,6)


DATA_PATH = "/content/drive/MyDrive/amazon-data-project/data/amazon.csv"
df = pd.read_csv(DATA_PATH, encoding="utf-8-sig")


def clean_money(series):
    return pd.to_numeric(
        series.astype(str)
              .str.replace(",", "", regex=False)
              .str.replace(r"[^\d.]", "", regex=True),
        errors="coerce"
    )

df["discounted_price"] = clean_money(df["discounted_price"])
df["actual_price"] = clean_money(df["actual_price"])

df["discount_percentage"] = pd.to_numeric(
    df["discount_percentage"].astype(str).str.replace("%", "", regex=False),
    errors="coerce"
)

df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df["rating_count"] = pd.to_numeric(
    df["rating_count"].astype(str).str.replace(",", "", regex=False),
    errors="coerce"
)

df["category"] = df["category"].astype(str).str.strip()

df = df.dropna(subset=[
    "discounted_price",
    "actual_price",
    "rating",
    "rating_count",
    "category"
])

df["discount_amount"] = df["actual_price"] - df["discounted_price"]

print("Data cleaning completed.")
print(df.info())
display(df.head())


OUTPUT_DIR = Path("/content/drive/MyDrive/amazon-data-project/outputs/figures")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

money = FuncFormatter(lambda x, pos: f"${x:,.0f}")


top_n = 10
avg_discount = (
    df.groupby("category")["discount_amount"]
      .mean()
      .sort_values(ascending=False)
      .head(top_n)
)

plt.figure()
ax = sns.barplot(x=avg_discount.values, y=avg_discount.index, orient="h")
ax.xaxis.set_major_formatter(money)
plt.title(f"Top {top_n} Categories by Average Discount")
plt.xlabel("Average Discount")
plt.ylabel("Category")

for i, v in enumerate(avg_discount.values):
    ax.text(v, i, f"  ${v:,.0f}", va="center")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "top_categories_avg_discount.png", dpi=300)
plt.show()


stats_cat = (
    df.groupby("category")["discount_amount"]
      .agg(mean="mean", median="median", count="size")
      .sort_values("mean", ascending=False)
      .head(top_n)
)

plot_df = (
    stats_cat.reset_index()
             .melt(id_vars="category",
                   value_vars=["mean", "median"],
                   var_name="metric",
                   value_name="discount_amount")
)

plt.figure()
ax = sns.barplot(data=plot_df,
                 x="discount_amount",
                 y="category",
                 hue="metric",
                 orient="h")

ax.xaxis.set_major_formatter(money)
plt.title(f"Top {top_n} Categories: Mean vs Median Discount")
plt.xlabel("Discount Amount")
plt.ylabel("Category")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "mean_vs_median_discount_by_category.png", dpi=300)
plt.show()


top_k = 8
top_categories = (
    df.groupby("category")["discount_amount"]
      .median()
      .sort_values(ascending=False)
      .head(top_k)
      .index
)

df_top = df[df["category"].isin(top_categories)].copy()

plt.figure()
ax = sns.boxplot(data=df_top,
                 x="discount_amount",
                 y="category",
                 showfliers=False)

ax.xaxis.set_major_formatter(money)
plt.title(f"Discount Distribution (Top {top_k} Categories)")
plt.xlabel("Discount Amount")
plt.ylabel("Category")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "discount_boxplot_top_categories.png", dpi=300)
plt.show()


plt.figure()
ax = sns.regplot(data=df,
                 x="discount_amount",
                 y="rating",
                 scatter_kws={"alpha":0.25, "s":12},
                 line_kws={"linewidth":2},
                 ci=95)

ax.xaxis.set_major_formatter(money)
plt.title("Rating vs Discount Amount")
plt.xlabel("Discount Amount")
plt.ylabel("Rating")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "rating_vs_discount_regression.png", dpi=300)
plt.show()


plt.figure()
ax = sns.histplot(df["discounted_price"], bins=40, kde=True)
ax.xaxis.set_major_formatter(money)
plt.title("Distribution of Discounted Prices")
plt.xlabel("Discounted Price")
plt.ylabel("Number of Products")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "discounted_price_distribution.png", dpi=300)
plt.show()


plt.figure()
sns.histplot(np.log1p(df["discounted_price"]), bins=40, kde=True)
plt.title("Distribution of Discounted Prices (log1p)")
plt.xlabel("log1p(Discounted Price)")
plt.ylabel("Number of Products")

plt.tight_layout()
plt.savefig(OUTPUT_DIR / "discounted_price_distribution_log.png", dpi=300)
plt.show()


print(f"\n Charts saved in:\n{OUTPUT_DIR}")
```
